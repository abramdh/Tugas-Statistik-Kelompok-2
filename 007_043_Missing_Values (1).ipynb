#Menggunakan Metode Imputation Dengan Menggantikan Nilai Missing Dengan Mean, Median dan Modus

````python
#Menampilkan jumlah missing values pada dataset
df = pd.read_csv('https://raw.githubusercontent.com/abramdh/datanarapidana/main/german_credit_data.csv')
print(df.isna().sum(), '\n\n',df.isna().any())
````

````python
#Mengambil dataset dari url, serta mengganti nilai pada missing values dengan nilai rata-rata atau mean pada dataset
df["Credit amount"] = df["Credit amount"].replace(np.NaN, df["Credit amount"].mean())
print(df.isna().sum(), '\n\n',df.isna().any())
````

````python
#Mengambil dataset dari url, serta mengganti nilai pada missing values dengan nilai tengah atau median pada dataset
df["Credit amount"] = df["Credit amount"].replace(np.NaN, df["Credit amount"].median())
print(df.isna().sum(), '\n\n',df.isna().any())
````

````python
#Mengganti nilai pada missing values dengan nilai terbanyak muncul atau modus
modus = df['Credit amount'].mode()[0]
df['Credit amount'] = df['Credit amount'].fillna(modus)
df['Credit amount'].isna().sum()
print(df.isna().sum(), '\n\n',df.isna().any())
````
